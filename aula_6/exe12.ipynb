```python
#Exercício 12 - Inner Join
#Objetivo: Unir os DataFrames de livros e empréstimos usando o método inner join e mostrar apenas os livros que foram emprestados.
```

```python
#Instruções:
#1.Carregue os arquivos livros.csv e emprestimos.csv em DataFrames separados.
```

```python
import pandas as pd
```

```python
livros = pd.read_csv('C:/Users/Pichau/Downloads/gc-python-2024/aula_6/livros.csv')
emprestimos = pd.read_csv('C:/Users/Pichau/Downloads/gc-python-2024/aula_6/emprestimos.csv')
```

```python
livros
```

```python
emprestimos
```

```python
#2. Realize um inner join entre as tabelas, unindo-as pela coluna id_livro.
```

```python
mesclagem = pd.merge(livros, emprestimos, on='id_livro', how='inner')
mesclagem.head(10)
```

```python
#3. Exiba o resultado, mostrando as colunas: titulo, autor, nome_leitor, data_emprestimo e data_devolucao.
```

```python
colunas_selecionadas = ['titulo', 'autor', 'nome_leitor', 'data_emprestimo', 'data_devolucao']
resultado_filtrado = mesclagem[colunas_selecionadas]

resultado_filtrado.head(10)
```
