```python
import pandas as pd
```

```python
filmes = ['Interestelar', 'The Batman', 'Avatar 2', 'Um lugar silencioso', 'Arcane']
data_lancamento = ['2014', '2022', '2022', '2024', '2021']
```

```python
filmes = pd.DataFrame ({
    'Nomes' : filmes,
    'Lançamento' : data_lancamento
})
```

```python
filmes
```

```python
filmes['Notas'] = [10, 10, 9, 10, 9]
filmes
```

```python
series = {
    'Nomes' : ['Breaking Bad', 'Peaky Blinders', 'The Midnight Gospel', 'Stranger Things', 'The End Of The Fucking World'],
    'Lançamento' : ['2008', '2013', '2020', '2016', '2017'],
    'Notas' : [10, 9, 10, 10, 10]
}

animes = {
    'Nomes' : ['Dragon Ball', 'One Piece', 'Naruto', 'Jujutsu Kaisen', 'Mob Psycho 100'],
    'Lançamento' : ['1986', '1997', '2002', '2018', '2016'],
    'Notas' : [10, 10, 10, 10, 10]
}
```

```python
series = pd.DataFrame(series)
animes = pd.DataFrame(animes)
```

```python
series
```

```python
animes
```

```python
filmes.to_csv('filmes.csv')
filmes.to_json('filmes.json')
filmes.to_html('filmes.html')
series.to_csv('series.csv')
series.to_json('series.json')
series.to_html('series.html')
animes.to_csv('animes.csv')
animes.to_json('animes.json')
animes.to_html('animes.html')
```

```python
df_concat = pd.concat([filmes, series, animes], axis=1)
```
