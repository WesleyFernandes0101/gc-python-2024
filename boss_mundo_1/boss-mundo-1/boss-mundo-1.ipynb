```python
pokedex.duplicated().sum()
```

```python
#Remover colunas irrelevantes:
#Analise as colunas presentes e exclua aquelas que não são relevantes para a análise.
```

```python
pokedex = pokedex.drop(['Unnamed: 0', 'german_name', 'japanese_name', 'base_friendship', 'base_experience', 'growth_rate', 'egg_type_number', 'egg_type_1', 'egg_type_2', 'percentage_male', 'egg_cycles'], axis=1)
pokedex
```

```python
#Tratar valores ausentes:
#Identifique valores nulos ou ausentes e aplique as ações necessárias para corrigi-los (preenchimento ou remoção).
```

```python
pokedex.isnull().sum()
```

```python
pokedex[['type_2', 'weight_kg', 'ability_1', 'ability_2', 'ability_hidden', 'catch_rate']] = pokedex[['type_2', 'weight_kg', 'ability_1', 'ability_2', 'ability_hidden', 'catch_rate']].fillna('Nothing assigned')
pokedex
```

```python
#Filtrar Pokémon proibidos:
#Exclua do conjunto de dados todos os Pokémon que pertencem às seguintes categorias:
# Lendários e Sub-lendários
# Míticos
# Pokémon com Mega Evoluções
```

```python
pokedex.drop(pokedex[pokedex['status'] != 'Normal'].index, inplace=True)
pokedex
```

```python
pokedex = pokedex[~pokedex['name'].str.startswith('Mega')]
pokedex
```
