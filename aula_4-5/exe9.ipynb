```python
filmes = pd.read_json('../aula_03/filmes.json')
series = pd.read_json('../aula_03/series.json')
animes = pd.read_json('../aula_03/animes.json')
```

```python
filmes2 = {
    'Nomes' : ['Oppenheimer', 'Kung Fu Panda', 'Sonic', 'Maze Runner', 'Blade Runner 2049'],
    'Lançamento' : [2023, 2008, 2020, 2014, 2017],
    'Notas' : [10, 7, 8, 10, 9]
}

series2 = {
    'Nomes': ['The Office', 'Sherlock', 'Friends', 'The Witcher', 'Black Mirror'],
    'Lançamento': [2005, 2010, 1994, 2019, 2011],
    'Notas': [8, 9, 8, 8, 8]
}

animes2 = {
    'Nomes': ['Fullmetal Alchemist', 'Your Name', 'Tokyo Ghoul', 'Koe No Katachi', 'Hunter x Hunter'],
    'Lançamento': [2003, 2016, 2014, 2017, 2011],
    'Notas': [9, 10, 7, 10, 10]
}
```

```python
filmes2 = pd.DataFrame(filmes2)
series2 = pd.DataFrame(series2)
animes2 = pd.DataFrame(animes2)
```

```python
concat_geral = pd.concat([filmes, series, animes, filmes2, series2, animes2], axis=0)
concat_geral
```

```python
concat_geral.iloc[:10]
```

```python
media_coluna = concat_geral['Notas'].mean()
print(f'O valor médio da primeira coluna é: {round(media_coluna, 1)}')
```

```python
concat_geral[concat_geral['Notas'] > media_coluna]
```
